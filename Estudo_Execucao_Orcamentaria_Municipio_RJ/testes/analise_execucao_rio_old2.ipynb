```python
# Agrupando os Orçamento por Exercicio e Orgão.
colunasGroupBy = [ 'Exercicio', 'Orgao']
colunas = ['Exercicio', 'Orgao','Orcamento_Atualizado']

df_total_exercicio_orgao = df_execucao[colunas].groupby(colunasGroupBy, as_index = False).agg({'Orcamento_Atualizado': 'sum'})

#print('Total por Exercíco e Órgão')
#df_total_exercicio_orgao
```

```python
# Define o tamanho do "Top" que será considerado.
top = 6

# Cria uma lista com todos os anos de exercício
anos_exercicio = df_total_exercicio_orgao['Exercicio'].unique()

# Cria uma lista de Dataframes de cada ano.
list_df = []

for ano in anos_exercicio:
    # Filtra os dados por ano.
    df_ano = df_total_exercicio_orgao[df_total_exercicio_orgao['Exercicio'] == ano]
   
    # Ordena os valores
    df_ano = df_ano.sort_values(by=['Orcamento_Atualizado'], ascending=False)

    # Converte a coluna Orgao para String
    df_ano['Orgao'] = df_ano['Orgao'].apply(str)
    
    # Pega os TOP maiores valores de orçamento
    df_top = df_ano[:top]
    
    # Calcula a adiciona a coluna outros.
    df_top_outros = df_top.reset_index(drop=True)
    df_top_outros.loc[len(df_top)] = [ano, 'Outros',df_ano[top:]['Orcamento_Atualizado'].sum()]
    
    list_df.append(df_top_outros)
```

```python
# Define o tamanho d figura do plot
fig = plt.figure(figsize=(20,20))

# Quantidade de linhas do plot
linhas_plot = 3

i = 1

# Plota os graficos do Orçamento por Orgão de cada ano.
for df in list_df:
    
    ax = fig.add_subplot(linhas_plot,len(list_df)/linhas_plot, i)
    #ax.axis('equal')
   
    # Dados do Plot
    df = df.sort_values(by=['Orgao'])
    dados = df['Orcamento_Atualizado'].values
    ano = df['Exercicio'].max()
    labels = df['Orgao'].unique()

    # Destada a fatia "Outros"
    explode = np.zeros(len(df))
    explode[len(explode)-1] = 0.1

    # Define o grafico
    plt.pie(dados , explode=explode,labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
    
    plt.title('Orçamento por Orgão: Top {0} em {1}'.format(top,  str(ano)))
    
    i+=1

plt.tight_layout()
plt.show()
```

```python
def agrupaValores(categoria_agrupar, valor_agrupar):
   
    colunasGroupBy = [ 'Exercicio', categoria_agrupar]
    colunas = ['Exercicio', categoria_agrupar, valor_agrupar]

    df_agrupado = df_execucao[colunas].groupby(colunasGroupBy, as_index = False).agg({valor_agrupar: 'sum'})

    return df_agrupado
    
```

```python
# Define o tamanho do "Top" que será considerado.
top = 6

def getCategoriaValorAnalise(df_agrupado, categoria, valor):

    # Cria uma lista com todos os anos de exercício
    anos_exercicio = df_agrupado['Exercicio'].unique()

    # Cria uma lista de Dataframes de cada ano.
    list_df = []

    for ano in anos_exercicio:
        # Filtra os dados por ano.
        df_ano = df_agrupado[df_agrupado['Exercicio'] == ano]

        # Ordena os valores
        df_ano = df_ano.sort_values(by=[valor], ascending=False)

        # Converte a coluna Orgao para String
        df_ano[categoria] = df_ano[categoria].apply(str)

        # Pega os TOP maiores valores de orçamento
        df_top = df_ano[:top]

        # Calcula a adiciona a coluna outros.
        df_top_outros = df_top.reset_index(drop=True)
        df_top_outros.loc[len(df_top)] = [ano, 'Outros',df_ano[top:][valor].sum()]

        list_df.append(df_top_outros)
    
    return list_df  
    
```

```python

def plotaGraficoPizza(lista_df, categoria_analise, valor_analise, top, titulo):
    
    # Define o tamanho d figura do plot
    fig = plt.figure(figsize=(20,20))

    # Quantidade de linhas do plot
    linhas_plot = 3

    i = 1

    # Plota os graficos do Orçamento por Orgão de cada ano.
    for df in lista_df:

        ax = fig.add_subplot(linhas_plot,len(lista_df)/linhas_plot, i)
        #ax.axis('equal')

        # Dados do Plot
        df = df.sort_values(by=[categoria_analise])
        dados = df[valor_analise].values
        ano = df['Exercicio'].max()
        labels = df[categoria_analise].unique()

        # Destada a fatia "Outros"
        explode = np.zeros(len(df))
        explode[len(explode)-1] = 0.1

        # Define o grafico
        plt.pie(dados , explode=explode,labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)

        plt.title('{0}: Top {1} em {2}'.format(titulo, top,  str(ano)))

        i+=1

    plt.tight_layout()
    plt.show()
```
