```python
#Plotando os valores por exercício e orgão

# Dados para plotagem
dados =  df_top_orgao['Orcamento_Atualizado'].values
label =  df_top_orgao['Orgao'].unique()

# Criando o plot
fig = plt.figure(figsize=(10,10))
ax = fig.subplots()

# Destada a fatia "Outros"
explode = np.zeros(len(df_top_orgao))
explode[len(explode)-1] = 0.1

#Define o grafico
ax.pie(dados , explode=explode,labels=label, shadow=True, startangle=0)

plt.title('Orçamento por Orgão: Top {0} em {1}'.format(top,  str(df_top_orgao['Exercicio'].max())))

plt.show()
```

```python
# Define o tamanho do "Top" que será considerado.
top = 7

# Monta uma lista com todos os anos de exercício
anos_exercicio = df_total_exercicio_orgao['Exercicio'].unique()

df_list = []

for ano in anos_exercicio:
    # Filtra os dados por ano.
    df_ano = df_total_exercicio_orgao[df_total_exercicio_orgao['Exercicio'] == ano]
   
    # Ordena os valores
    df_ano = df_ano.sort_values(by=['Orcamento_Atualizado'], ascending=False)

    # Converte a coluna Orgao para String
    df_ano['Orgao'] = df_ano['Orgao'].apply(str)
    
    # Pega os TOP maiores valores de orçamento
    df_top = df_ano[:top]
    
    # Calcula a adiciona a coluna outros.
    df_top_outros = df_top.reset_index(drop=True)
    df_top_outros.loc[len(df_top)] = [ano, 'Outros',df_ano[top:]['Orcamento_Atualizado'].sum()]
    
    #plotaGraficoPizzaOrcamento( df_top_outros['Orcamento_Atualizado'].values, df_top_orgao['Orgao'].unique(), df_top_outros['Exercicio'].max(), top)
    df_list.append(df_top_outros)
```

```python
#Plotando os valores por exercício e orgão
def plotaGraficoPizzaOrcamento(dados, labels, ano, top):
    
     
    # Criando o plot
    fig = plt.figure(figsize=(10,10))
    ax = fig.subplots()
    
    # Destada a fatia "Outros"
    explode = np.zeros(len(dados))
    explode[len(explode)-1] = 0.1

    #Define o grafico
    ax.pie(dados , explode=explode,labels=labels, autopct='%1.1f%%', shadow=True, startangle=0)

    plt.title('Orçamento por Orgão: Top {0} em {1}'.format(top,  str(ano)))

    plt.show()
```

```python
### Define algumas funções para ajudar no tratamento dos dados.

# Define a função que Remove Simbolo de moeda
def removeSimboloMoeda(string):
    return string.strip().replace('R$','')


#Define a função que Inverte o separador decimal.
def inverteSeparadorDecimal(string):
    string = str(string).replace(',','v')
    string = string.replace('.',',')
    string = string.replace('v','.')
    
    return string

#Define a função que Converte um objeto para float
def objToFloat(obj):
    return  locale.atof(removeSimboloMoeda(obj))


#Define a função que retorna os segundos no formato hh:mm:ss
def formata_segundos_hhmmss(segundos):
    return time.strftime('%H:%M:%S', time.gmtime(segundos))

#Define a função que retorna data atual do sistema.
def get_data_atual():
    return time.strftime("%d/%m/%Y")

#Define a função que retorna a data e hora atual do sistema.
def get_data_hora_atual():
    return time.strftime("%d/%m/%Y %H:%M:%S")

# Define a Função para Cria um diretório, se não existir
def cria_diretorio(path, recursivo=False):
    if not os.path.exists(path):
        if recursivo:
            os.makedirs(path)
        else:
            os.mkdir(path)
```
